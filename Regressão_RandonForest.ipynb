# REGRESSÃO COM RANDOM FOREST

```python
independente_rf = df.iloc[:, 3:].values
independente_rf
```

```python
dependente_rf = df.iloc[:, 1].values
dependente_rf
```

```python
independente_rf.shape
```

```python
dependente_rf.shape
```

### Modelo

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(independente_rf, dependente_rf, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape, y_treino.shape
```

```python
x_teste.shape, y_teste.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random = RandomForestRegressor(n_estimators=60, criterion='squared_error', max_depth=4, random_state = 10)
random.fit(x_treino, y_treino)
```

```python
random.score(x_treino, y_treino)
```

```python
random.score(x_teste, y_teste)
```

```python
previsoes_teste = random.predict(x_teste)
previsoes_teste
```

```python
random.n_features_in_
```

### Métricas

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Erro médio Absoluto (MAE)
mean_absolute_error(previsoes_teste, y_teste)
```

```python
# Erro médio Quadratico (MSE)
mean_squared_error(previsoes_teste, y_teste)
```

```python
# Raiz do Erro médio Quadratico (RMSE)
np.sqrt(mean_squared_error(previsoes_teste, y_teste))
```
